# Step 2: each country leagues:

```python
import pandas as pd
coutries_dict={}
countires=['England','Germany','Italy','France','spain']
links=['https://www.transfermarkt.com/wettbewerbe/national/wettbewerbe/189',
       'https://www.transfermarkt.com/wettbewerbe/national/wettbewerbe/40',
       'https://www.transfermarkt.com/wettbewerbe/national/wettbewerbe/75',
       'https://www.transfermarkt.com/wettbewerbe/national/wettbewerbe/50',
       'https://www.transfermarkt.com/wettbewerbe/national/wettbewerbe/157']
codes=[189,40,75,50,157] 
coutries_dict['Country']=countires
coutries_dict['Code']=codes
coutries_dict['URL_link']=links
df=pd.DataFrame(data=coutries_dict)
df
```

```python
#creating data link for each season
import numpy as np 
years=np.arange(2015,2022)
for link in df['URL_link']:
    for year in years:
       season_link=link+'/saison_id/'+ str(year)
       season_link_detailed=link+'/plus/1?saison_id='+ str(year)
       print(season_link)
       print(season_link_detailed)
```
