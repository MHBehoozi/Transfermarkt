```python

import requests
from bs4 import BeautifulSoup
from lxml import etree
import pandas as pd
```

```python
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}
club_values=[]
coach=[]
coach_date=[]
href_name=[]
    
```

### Premier League

```python

def get_premier_league_teams(url):

    response = requests.get(url=url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('div', class_='responsive-table')

    teams = []
    
    

    if table:
        rows = table.find_all('td')
        
        for row in rows:
            team_link = row.find('a')
            
            if team_link:
                team_href = team_link.get('href')
                if team_href not in teams and 'startseite'in team_href:
                   teams.append(team_href)
                   href_name.append(team_href.split("/")[1])
                   
    
    
    return teams
```

```python
def coach_name(soup):
    box=soup.find_all('div', attrs={'data-viewport':'Mitarbeiter_Historie'})[0]
    name=box.find('ul').find('li').find('div',attrs={'class':'container-main'}).find('a').text
    date=box.find('ul').find('li').find('div',attrs={'class':'container-tenure'}).text
    date=date.split('\n')[1].split('\xa0â€“\xa0')[0].rstrip().lstrip()
    if name not in coach:
        coach.append(name)
        coach_date.append(date)
    return 
```

```python
def market_value(soup):
    value = soup.select('#main > main > header > div.data-header__box--small > a')
    
    try:
        current_value=value[0].text.strip().replace('Total market value','').replace('\n','').strip()
        if current_value not in club_values:
           club_values.append(current_value)
           return
    except:
        club_values.append('NULL')
        return
```

```python
def team_link(suburl):
    url = f'https://www.transfermarkt.com{suburl}'
    response = requests.get(url=url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    name = soup.select('#main > main > header > div.data-header__headline-container > h1')
    league_name = soup.select('#main > main > header > div.data-header__box--big > div > span.data-header__club > a')
    #coach=coach_name()
    avg_palyers=soup.select('#main > main > header > div.data-header__info-box > div > ul:nth-child(1) > li:nth-child(2)')
    market_value(soup)
    coach_name(soup)
    avg=avg_palyers[0].text.rstrip('\n').strip()
    try:
       stadium_name=soup.select('#main > main > header > div.data-header__info-box > div > ul:nth-child(2) > li:nth-child(2) > span > a')
       s=stadium_name[0].text
    except:
        s='NULL'
    try:
        n=name[0].text.strip()
    except:
        n='NULL'
    try:
        l=league_name[0].text.strip()
    except:
        l='NULL'        

   
       
    return n,l,avg.replace('\n','').replace('Average age:','').strip(),s
```

```python


for id_year in ['2021','2020','2019','2018','2017','2016','2015']:
    club_values=[]
    coach=[]
    coach_date=[]
    href_name=[]
    data_dict=[]
    club_name=[]
    stadium_name=[]
    league_name=[]
    avg_players=[]
    url=f'https://www.transfermarkt.com/premier-league/startseite/wettbewerb/GB1/plus/?saison_id={id_year}'
    premier_league_teams = get_premier_league_teams(url)
    premier_league_teams=list(dict.fromkeys(premier_league_teams))
    for href in premier_league_teams:
        club,league,avg_player,stadium=team_link(href)
        club_name.append(club)
        league_name.append(league)
        avg_players.append(avg_player)
        stadium_name.append(stadium)
        data_dict.append(
       {
        'Club Name':club_name.pop(),
        'League name':league_name.pop(),
        'Average age of players':avg_players.pop(),
        'Stadium name':stadium_name.pop(),
        'Coach name':coach.pop(),
        'Coach Date':coach_date.pop(),
        'Club Values': club_values.pop(),
       }
       ) 
    df = pd.DataFrame(data_dict,columns=['Club Name',
        'League name',
        'Average age of players',
        'Stadium name',
        'Coach name',
        'Coach Date',
        'Club Values'])
    df.to_csv(f'PriemerLeague{id_year}.csv')     
```
