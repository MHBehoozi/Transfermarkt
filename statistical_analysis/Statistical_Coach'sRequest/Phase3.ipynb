```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('palyers_statistic .csv')
df
```

```python
position_df=pd.read_csv('palyers.csv')
position_df
```

```python
market_values=pd.read_csv('market_value_df.csv')
market_values
```

```python
in_df=pd.merge(df,position_df,on='player_id',how='left')
main_df=pd.merge(in_df,market_values,on='player_id',how='left')
main_df
```

```python
Forward_df=main_df[main_df['main_position'].isin(['Attack','Attack Right Winger','Attack Centre Forward','Attack Left Winger'])]
Forward_df
```

```python
Goalkeeper_df=main_df[main_df['main_position']=='Goalkeeper']
Goalkeeper_df
```

```python

metric_score=[]
playe_id=Forward_df['player_id']
goals=Forward_df['Goals']
assist=Forward_df['Assists']

df['Clean sheets']
```

```python
import math
metric_score=[]
for index,record in main_df.iterrows():
    if record['main_position'] in ['Attack','Attack Right Winger','Attack Centre Forward','Attack Left Winger']:
       goal=int(record['Goals'])
       Assist=int(record['Assists'])
       forward_metric=(3*goal+Assist)
       metric_score.append(forward_metric)
    elif record['main_position'] =='Goalkeeper':
        clean_sheet=int(record['Clean sheets'])
        # Assist=int(record['Assists'])
        Goalkeeper_metric=(5*clean_sheet)
        metric_score.append(Goalkeeper_metric)
    elif record['main_position'] in ['Defender Centre Back','Defender Left Back','Defender Right Back']:
        if math.isnan(record['Clean sheets']):
           clean_sheet=0
        else:
            clean_sheet=int(record['Clean sheets'])   
        red_card=int(record['Red cards'])
        Minutes=int(record['Minutes played'])
        Defender_metric=(4*clean_sheet-red_card)
        if Minutes >= 100:
            increment = Minutes // 100
            Defender_metric += increment+1
        # Assist=int(record['Assists'])
        metric_score.append(Defender_metric)
    elif record['main_position'] in ['midfield Defensive Midfield','midfield Central Midfield','midfield Attacking Midfield','midfield Left Midfield','midfield Right Midfield']:
        goal=int(record['Goals'])
        red_card=int(record['Red cards'])
        Assist=int(record['Assists'])
        midfield_metric=((2*goal+3*Assist)-red_card)
        Minutes=int(record['Minutes played'])
        if Minutes >= 100:
            increment = Minutes // 100
            midfield_metric += increment+1
        metric_score.append(midfield_metric)    
    else:
         metric_score.append(0)   

metric_score      
```

```python
main_df['Score']=metric_score
main_df
```

```python
s_df=main_df[main_df['season_x']==2021]
final_df=s_df.groupby(['player_id','market_value_y'])['Score'].sum().reset_index()
final_df
```

```python
# محاسبه رتبه براساس عملکرد
final_df['Score_Rank'] = final_df['Score'].rank(ascending=False)
final_df['market_value_Rank'] = final_df['market_value_y'].rank(ascending=True)
final_df
```

```python
df_sorted = final_df.sort_values(by=['player_id', 'market_value_y'], ascending=False)

final_df_sorted = df_sorted.drop_duplicates(subset='player_id', keep='first')
final_df_sorted
```

```python
final_df_sorted['عملکرد درصد'] = final_df_sorted['Score_Rank'] / len(df)

# محاسبه درصد رتبه قیمت
final_df_sorted['قیمت درصد'] =final_df_sorted['market_value_Rank'] / len(df)
final_df_sorted
```

```python
# Assuming you have a DataFrame called df and you want to calculate the intersection
# based on the top 30% performance and the bottom 40% price

# Calculate the cutoff values for performance and price
dl=final_df_sorted.sort_values(by='عملکرد درصد',)
performance_cutoff = dl['عملکرد درصد'].quantile(0.7)  # Top 30%
dl=final_df_sorted.sort_values(by='قیمت درصد')
price_cutoff = dl['قیمت درصد'].quantile(0.6)  # Bottom 40%

# Create the subsets based on the cutoff values
bottom_price_subset = dl[final_df_sorted['قیمت درصد'] <= price_cutoff]
dl=final_df_sorted.sort_values(by='عملکرد درصد')
top_performance_subset = dl[final_df_sorted['عملکرد درصد'] >= performance_cutoff]


# Calculate the intersection
intersection_df = final_df_sorted.loc[top_performance_subset.index.intersection(bottom_price_subset.index)]

# Print the intersection
intersection_df
```

```python
final_df_sorted['عملکرد درصد'] = final_df_sorted['Score_Rank'] / len(df)

# محاسبه درصد رتبه قیمت
final_df_sorted['قیمت درصد'] =final_df_sorted['market_value_Rank'] / len(df)

# فیلتر کردن براساس شرایط مورد نظر (30% برترین عملکرد و 40% پایین‌ترین قیمت)
filtered_df = final_df_sorted[(final_df_sorted['عملکرد درصد'] <= 0.3) & (final_df_sorted['قیمت درصد'] <= 0.4)]
dl=filtered_df.sort_values(by='عملکرد درصد')
dl
```
