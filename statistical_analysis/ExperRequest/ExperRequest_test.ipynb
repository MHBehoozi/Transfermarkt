```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N = 200
df = pd.DataFrame({ 'playerID' : range(1, N + 1, 1),
    'MV' : np.random.randint( 1000000, 1550000, size=N),
    'fee' : np.random.randint(1150000, 1700000, size=N)})
df.head()
```

```python
df = df.sort_values(by=['fee', 'MV'])
```

```python
X_axis = np.arange(len(df['playerID']))
plt.figure(figsize = (10,5))
plt.bar(X_axis - 0.2, df['MV']/1000000, 0.4, label = 'MV')
plt.bar(X_axis + 0.2, df['fee']/1000000, 0.4, label = 'fee')
plt.xticks(X_axis, df['playerID'])
plt.xlabel("playerID")
plt.ylabel("million dollar")
plt.legend() 
```

```python
kwargs = dict(alpha=0.5, bins=10000, density=True, stacked=False)
 
# Plotting the histogram
plt.hist(df['MV'],histtype='stepfilled',color='b',label='MV')
plt.hist(df['fee'],histtype='stepfilled',color='r',label='fee')

plt.gca().set(title='Histogram', ylabel='Probability')

plt.legend()
```

```python
bins = np.linspace(-100, 100, 1000)
plt.hist(df['MV'],bins=100, alpha=0.5, label='MV', color='r')
plt.hist(df['fee'], bins=100, alpha=0.5, label='fee', color='g')
plt.legend(loc='upper right')
plt.show()
```

```python
bins = np.linspace(-100, 100, 1000)
plt.hist([10,12,13,14],bins=100, alpha=0.5, label='MV', color='r')
plt.hist([12,13,14,15], bins=100, alpha=0.5, label='fee', color='g')
plt.legend(loc='upper right')
plt.show()
```
