# import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sqlalchemy import create_engine
from sqlalchemy import URL
from sqlalchemy import text
from urllib.parse import quote_plus
```

# create sql URL

```python
url_object = URL.create(
    "mysql+mysqlconnector",
    username="root",
    password="",
    host="localhost"
    )
```

# create sengine

```python
engin = create_engine(url_object)
```

# Fetch the required DATA

```python
query = ("SELECT player_id, MV, transfer_Fee, season FROM player_transfers "
         "WHERE "
         "transfer_Fee != 0 "
         "and "
         "MV != 0;") 
```

```python
with engin.connect() as conn:
    conn.execute(text("USE quera_project1"))
    df = pd.read_sql(text(query), conn)
```

```python
df.head()
```

# Distribution chart by season

# 2017

```python
df17 = df[df['season'] == 2017]
```

```python
# create x data to compute probiblity
xmin, xmax = 0, 60000000
x = np.linspace(xmin, xmax, 100)
```

```python
fig, ax = plt.subplots()
p1 = stats.norm.pdf(x, np.mean(df17['MV']),  np.std(df17['MV']))
p2 = stats.norm.pdf(x, np.mean(df17['transfer_Fee']), np.std(df17['transfer_Fee']))
ax.plot(x, p1, 'b', label = 'Estimated Value (MV)')
ax.plot(x, p2, 'r', label = 'Real Market Value (Fee)')
ax.axvline(np.mean(df17['MV']), color = 'b', linestyle='--')
ax.axvline(np.mean(df17['transfer_Fee']), color = 'r', linestyle='--')
ax.set_title('Comparison of real value vs estimated value of players\n(normal distribution)')
ax.legend()
ax.set_xlabel('Market Value')

difference_value = round((np.mean(df17['transfer_Fee']) - np.mean(df17['MV']))/1000000,3)
fig.text(0.49,0.52,('Difference between average of\nreal value and estimated value\nis '+str(difference_value)+'*1e6'))
```
