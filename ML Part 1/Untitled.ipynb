```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from jupyter_datatables import init_datatables_mode
```

```python
defender_df = pd.read_csv('defender_df.csv')
Goalkeeper_df = pd.read_csv('Goalkeeper_df.csv')
midfielder_df = pd.read_csv('midfielder_df.csv')
striker_df = pd.read_csv('striker_df.csv')
```

```python
striker_df.head()
```

```python
X_striker = striker_df.drop(['market_value', 'Competition', 'position_group','Club', 'club_id'], axis=1)
y_striker = striker_df['market_value']
```

```python
# فرض میکنیم داده‌ها را در متغیرهای 'data' و برچسب‌ها را در متغیرهای 'labels' دارید
# تقسیم داده‌ها به دو بخش اصلی آموزش و تست
train_data, test_data, train_labels, test_labels = train_test_split(X_striker, y_striker, test_size=0.2, random_state=42)

# تقسیم داده‌های آموزشی به دو بخش آموزش و اعتبارسنجی
train_data, val_data, train_labels, val_labels = train_test_split(train_data, train_labels, test_size=0.125, random_state=42)

# چاپ ابعاد مجموعه داده‌های حاصل
print("ابعاد داده‌های آموزش:", train_data.shape)
print("ابعاد برچسب‌های آموزش:", train_labels.shape)
print("ابعاد داده‌های اعتبارسنجی:", val_data.shape)
print("ابعاد برچسب‌های اعتبارسنجی:", val_labels.shape)
print("ابعاد داده‌های تست:", test_data.shape)
print("ابعاد برچسب‌های تست:", test_labels.shape)
```

```python
# ایجاد شیء مدل رگرسیون خطی
model = LinearRegression()

# آموزش مدل با داده‌های آموزش
model.fit(train_data, train_labels)

# پیش‌بینی برچسب‌ها برای داده‌های اعتبارسنجی
val_predictions = model.predict(val_data)

# پیش‌بینی برچسب‌ها برای داده‌های تست
test_predictions = model.predict(test_data)

# محاسبه دقت مدل بر روی داده‌های اعتبارسنجی و تست (مثال: ضریب تعیین)
val_accuracy = model.score(val_data, val_labels)
test_accuracy = model.score(test_data, test_labels)

# چاپ دقت مدل
print("دقت مدل بر روی داده‌های اعتبار")
print(val_accuracy)
print(test_accuracy)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# فرض کنید داده‌های آموزشی و برچسب‌ها را دارید

# تقسیم داده‌ها به سه قسمت: آموزشی، اعتبارسنجی و آزمون

# ساخت مدل گرادیان جنگل تصادفی با پارامترهای دلخواه
model = RandomForestRegressor(n_estimators=100, random_state=42)

# آموزش مدل با داده‌های آموزشی
model.fit(X_train, y_train)

# پیش‌بینی برچسب‌ها برای داده‌های اعتبارسنجی و آزمون
y_val_pred = model.predict(X_val)
y_test_pred = model.predict(X_test)

# محاسبه معیار میانگین خطا (MSE) برای داده‌های اعتبارسنجی و آزمون
val_mse = mean_squared_error(y_val, y_val_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

val_accuracy = model.score(val_data, val_labels)
test_accuracy = model.score(test_data, test_labels)

print("Validation Mean Squared Error:", val_mse)
print("Test Mean Squared Error:", test_mse)

print(val_accuracy)
print(test_accuracy)
```
